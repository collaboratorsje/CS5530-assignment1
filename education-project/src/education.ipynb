3. Data Analysis

```python
import pandas as pd

df = pd.read_csv('../data-raw/StudentsPerformance.csv')
```

Histogram of Students' Scores:

This will give a distribution of students' scores, helping to understand the overall performance of students.

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(df['writing score'])
plt.title('Distribution of Students\' Scores')
plt.show()
```

Boxplot of Scores by Group:

If the dataset contains group/class information, this will help identify the spread and outliers within each group.

```python
sns.boxplot(x='lunch', y='math score', data=df)
plt.title('Boxplot of Scores by Group')
plt.show()
```

Scatter Plot of Two Related Attributes:

Let's assume there are two columns (e.g., study time and scores) that might have a relationship. A scatter plot can visually show this relationship.

```python
sns.scatterplot(x='test preparation course', y='reading score', data=df)
plt.title('Scatter Plot of Scores vs completion of Test Prep Course')
plt.show()
```

Bar Plot of Average Scores by Category:

For a categorical variable (like gender or ethnicity), you can compare average scores across these categories.

```python
avg_scores = df.groupby('gender')['math score'].mean()
avg_scores.plot(kind='bar')
plt.title('Average Math Scores by Gender')
plt.show()
```

Violin Plot of Writing Scores by Test Preparation Course:

This allows us to see the distribution of writing scores based on whether a student completed a test preparation course or not.

```python
sns.violinplot(x='test preparation course', y='writing score', data=df)
plt.title('Writing Scores by Test Preparation Course')
plt.show()
```
